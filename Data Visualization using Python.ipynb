# Data Visualisation Using Python Seaborn

```python
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
```

```python
data = pd.read_csv('Weather_data.csv')
```

```python
data.head()  #Checking the top entries of the dataset
```

```python
data.info() #To get more information on the dataset: dtype and null values
```

## Univariate visualisations in seaborn

### Histogram

```python
sns.histplot(x = data["temperature"]) 
```

```python
sns.histplot(y='temperature',data=data)  ##switching the axis
```

```python
sns.histplot(x = data["temperature"], hue=data['weather_type']) ## checking the weather types for temperatures
```

```python
sns.histplot(x = data["humidity"]) 
```

```python
sns.histplot(x = data["humidity"], hue=data['weather_type'])
```

### Distplot

```python
sns.distplot(data["temperature"]) 
```

```python
sns.distplot(data["humidity"]) 
```

### Displot

```python
sns.displot(data=data, x="temperature", kde=True) ##kde stands for kernel density estimation.
```

### Box plot

```python
sns.boxplot(x= data["temperature"]) 
```

```python
sns.boxplot(x= data["humidity"])
```

### Countplot

```python
sns.countplot(x =data['weather_type'])
```

```python
sns.countplot(y =data['weather_description'])
```

## Bivariate visualisations in seaborn

### Scatterplot

```python
sns.scatterplot(x=data["humidity"],y =data['temperature']) 
```

```python
sns.scatterplot(x=data["air_pollution_index"],y =data['temperature']) 
```

```python
sns.scatterplot(x=data["visibility_in_miles"],y =data['temperature']) 
```

### Lineplot

```python
sns.lineplot(x=data["humidity"],y =data['temperature']) 
```

```python
sns.lineplot(x=data["air_pollution_index"],y =data['temperature']) 
```

```python
sns.lineplot(x=data["visibility_in_miles"],y =data['temperature'])
```

### Regplot

```python
sns.regplot(x=data["humidity"],y =data['temperature']) 
```

```python
sns.regplot(x=data["air_pollution_index"],y =data['temperature']) 
```

```python
sns.regplot(x=data["visibility_in_miles"],y =data['temperature']) 
```

### Barplot (for categorial variables)

```python
sns.barplot(x= data['weather_type'], y = data['temperature'])
```

```python
sns.barplot(x= data['weather_type'], y = data['humidity'])
```

```python
sns.barplot(x= data['weather_type'], y = data['visibility_in_miles'])
```

```python
sns.barplot(x= data['weather_type'], y = data['temperature'], ci = 'sd')
```

```python
sns.barplot(x= data['weather_type'], y = data['humidity'], estimator = min)
```

### Pointplot

```python
sns.pointplot(x=data['weather_type'],y=data['temperature'])
```

```python
sns.pointplot(x=data['weather_type'],y=data['visibility_in_miles'])
```

### Jointplot

```python
sns.jointplot(x = data['humidity'],y = data['temperature'])
```

```python
sns.jointplot(x = data['humidity'],y = data['temperature'], kind='hex')
```

```python
sns.jointplot(x = data['humidity'],y = data['temperature'], kind='kde')
```

### Pairplot

```python
sns.pairplot(data[['humidity', 'temperature', 'air_pollution_index']])
```
